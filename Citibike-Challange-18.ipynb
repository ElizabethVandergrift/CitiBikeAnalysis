```python
import pandas as pd
from pathlib import Path
```

```python



# File Path
YR23_data_to_load = Path("JC-202303-citibike-tripdata.csv")
YR24_data_to_load = Path("JC-202403-citibike-tripdata.csv")

# Read Hospital General, Score, and Location Data File and store into Pandas DataFrames
YR23_data = pd.read_csv(YR23_data_to_load)
YR24_data = pd.read_csv(YR24_data_to_load)


YR23_data.head() 
```

```python
YR24_data.head()
```

```python

# Add a year column
YR23_data['year'] = 2023
YR24_data['year'] = 2024

# Combine DataFrames
combined_df = pd.concat([YR23_data, YR24_data])

# Save to CSV
combined_df.to_csv("combined_citibike_data.csv", index=False)

combined_df.head()
```
